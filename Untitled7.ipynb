```python
# Statistical Modeling: IMD and Obesity Association
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats


# 1. Prepare the data (assuming merged DataFrame exists from previous steps)
# Changed 'lsoa_boundaries7CD' to 'OtherColumn' as it's the likely intended column
model_data = merged[['OtherColumn', 'IMD Score', 'Value']].dropna()
model_data = model_data.rename(columns={'Value': 'Obesity_Pct'})

# Add quadratic term for potential non-linear relationship
model_data['IMD_Score_Squared'] = model_data['IMD Score']**2



# 2. Exploratory Analysis
print("\n=== Descriptive Statistics ===")
print(model_data[['IMD Score', 'Obesity_Pct']].describe())

print("\n=== Correlation ===")
corr, p_value = stats.pearsonr(model_data['IMD Score'], model_data['Obesity_Pct'])
print(f"Pearson r: {corr:.3f}, p-value: {p_value:.4f}")

# Scatterplot with regression line
plt.figure(figsize=(10,6))
sns.regplot(x='IMD Score', y='Obesity_Pct', data=model_data,
            scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title("Deprivation vs Obesity Prevalence (n={})".format(len(model_data)))
plt.xlabel("Index of Multiple Deprivation Score")
plt.ylabel("Obesity Prevalence (%)")
plt.show()

# 3. Linear Regression Model
X = sm.add_constant(model_data['IMD Score'])  # Add intercept
y = model_data['Obesity_Pct']

model = sm.OLS(y, X).fit()
print("\n=== Simple Linear Regression ===")
print(model.summary())

# 4. Polynomial Regression (testing non-linearity)
X_poly = sm.add_constant(pd.DataFrame({
    'IMD_Score': model_data['IMD Score'],
    'IMD_Score_Squared': model_data['IMD_Score_Squared']
}))

poly_model = sm.OLS(y, X_poly).fit()
print("\n=== Polynomial Regression ===")
print(poly_model.summary())

# 5. Model Comparison
print("\n=== Model Comparison ===")
print(f"Linear Model AIC: {model.aic:.2f}")
print(f"Polynomial Model AIC: {poly_model.aic:.2f}")

# 6. Diagnostic Plots
fig, axes = plt.subplots(1, 2, figsize=(12,5))

# Residuals plot
axes[0].scatter(model.fittedvalues, model.resid, alpha=0.5)
axes[0].axhline(y=0, color='r', linestyle='--')
axes[0].set_xlabel("Fitted values")
axes[0].set_ylabel("Residuals")
axes[0].set_title("Residuals vs Fitted")

# Q-Q plot
sm.qqplot(model.resid, line='s', ax=axes[1])
axes[1].set_title("Q-Q Plot")

plt.tight_layout()
plt.show()

```

# *Spatial Plot*  - Example 1

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
from shapely.geometry import Polygon

# 1. Create sample LSOA boundaries (polygons for 3 areas)
sample_geometries = [
    Polygon([(0,0), (1,0), (1,1), (0,1)]),
    Polygon([(1,1), (2,1), (2,2), (1,2)]),
    Polygon([(0,2), (1,2), (1,3), (0,3)])
]
lsoa = gpd.GeoDataFrame({
    'LSOA21CD': ['E01000001', 'E01000002', 'E01000003'],
    'geometry': sample_geometries
})

# 2. Create sample attribute data
merged = pd.DataFrame({
    'LSOA21CD': ['E01000001', 'E01000002', 'E01000003'],
    'IMD_Score': [15.2, 28.7, 42.1],
    'Obesity_Pct': [22.5, 31.6, 38.2]
})

# 3. Merge the data
gdf = lsoa.merge(merged, on='LSOA21CD')

# 4. Plot
fig, ax = plt.subplots(figsize=(10, 6))
gdf.plot(column='IMD_Score',
        ax=ax,
        legend=True,
        cmap='viridis',
        edgecolor='black',
        linewidth=0.5)
ax.set_title("Sample Deprivation Visualization")
ax.set_axis_off()
plt.show()
```
